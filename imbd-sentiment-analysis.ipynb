# **TEXT PREPROCESSING **

```python
import re
clean=re.compile('<.*?>')
def cleanhtml(raw):
  cleantext=re.sub(clean,'',raw)
  return cleantext
```

```python
df['review']=df['review'].apply(cleanhtml)
```

```python
#convert the review in lower case
df['review']=df['review'].apply(lambda x:x.lower())
```

```python
df['review'][0]
```

```python
#to remove special character of reviews
def removeschar(text):
    x=''
    for i in text:
        if i.isalnum():
          x=x+i
        else:
          x=x+' '
    return x
```

```python
df['review']=df['review'].apply(removeschar)
```

```python
df['review'][0]
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
def removestopwords(text):
    return " ".join([word for word in text.split()
                     if word not in stopwords.words('english')])
```

```python
removestopwords('the aryan is a very good boy')
```

```python
df['review']=df['review'].apply(removestopwords)
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stemming(text):
  return " ".join([ps.stem(word)for word in text.split()])
```

```python
df.head()
```

```python
df['review']=df['review'].apply(stemming)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000)
```

```python
x=cv.fit_transform(df['review']).toarray()
```

```python
y=df.iloc[:,-1].values
```

```python
y.shape
```

**MODEL TRAINING**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
gb=GaussianNB()
mb=MultinomialNB()
bb=BernoulliNB()
```

```python
clf1=gb.fit(X_train,y_train)
clf2=mb.fit(X_train,y_train)
clf3=bb.fit(X_train,y_train)
```

```python
y_pred1=gb.predict(X_test)
y_pred2=mb.predict(X_test)
y_pred3=bb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("gaussian acc=",accuracy_score(y_test,y_pred1))
print("multinomial acc=",accuracy_score(y_test,y_pred2))
print("bernoulli acc=",accuracy_score(y_test,y_pred3))
```

```python
import joblib
joblib.dump(mb,'model.pkl')
joblib.dump(cv,'countvectorizer.pkl')
```

```python
def preprocess(text):
    text = text.lower()
    text = cleanhtml(text)
    text = removeschar(text)
    text = removestopwords(text)
    text = stemming(text)
    return text
```

```python
preprocess('the movie was so good.i played yesterday and also will go to play today')
```

```python
review = input("Enter review: ")

review = preprocess(review)

vector = cv.transform([review])

prediction = mb.predict(vector)[0]

if prediction == 1:
    print("Positive 😊")
else:
    print("Negative 😞")
```
